```python
import pandas as pd
import numpy as np
import requests
import json
from pprint import pprint
import os 
import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
url = "https://statsapi.web.nhl.com/api/v1/teams"
```

```python
response = requests.get(url + "?expand=team.stats")
print(response.content)
```

```python
response = requests.get(url + "?expand=team.stats").json()
pprint(response)
```

```python
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
teams = response["teams"]
pprint(teams)
```

```python
teams = response["teams"]
goalspergame  =[]
gapg = []
for x in teams:
    output1=x["teamStats"]
    output2 = output1[0]["splits"]
    output3 = output2[0]["stat"]
    goalspergame.append(output3["goalsPerGame"])
    gapg.append(output3["goalsAgainstPerGame"])
print(goalspergame, gapg)
```

```python
df = pd.DataFrame({"goals per game": goalspergame, "goals against per game": gapg})
df
```

```python
df.plot.scatter(x = "goals per game", y = "goals against per game")
x = np.array(["goals per game"])
y = np.array(["goals against per game"])
# m, b = np.polyfit(x, y, 1)
# plt.plot(x, m*x +b)
```

```python
teams = response["teams"]
goalspergame  =[]
shotsallowed = []
for x in teams:
    output1 = x["teamStats"]
    output2 = output1[0]["splits"]
    output3 = output2[0]["stat"]
    goalspergame.append(output3["goalsPerGame"])
    shotsallowed.append(output3["shotsAllowed"])
print(goalspergame, shotsallowed)
```

```python
df = pd.DataFrame({"goals per game": goalspergame, "shots allowed per game": shotsallowed})
df
```

```python
df.plot.scatter(x = "goals per game", y = "shots allowed per game")
x = np.array(["goals per game"])
y = np.array(["shots allowed per game"])
```

```python
teams = response["teams"]
moreshots  =[]
outshot = []
for x in teams:
    output1 = x["teamStats"]
    output2 = output1[0]["splits"]
    output3 = output2[0]["stat"]
    moreshots.append(output3["winOutshootOpp"])
    outshot.append(output3["winOutshotByOpp"])
print(moreshots, outshot)
```

```python
df = pd.DataFrame({"more shots": moreshots, "outshot": outshot})
df
```

```python
df.plot.scatter(x = "more shots", y = "outshot")
x = np.array(["more shots"])
y = np.array(["outshot"])
```
